```python
data=pd.read_csv('C:/Users/ph871/Downloads/regression.csv')
```

```python
x=data.iloc[:,0].values
y=data.iloc[:,1].values
```

```python
data.corr()
```

```python
theta=[0,0]
eta=0.01
count = 200
```

```python
def gradDesc(theta):  
    a=(theta[0]+x*theta[1])-y
    b=x*a
    theta[0]=theta[0]-eta*sum(a)
    theta[1]=theta[1]-eta*sum(b)
    return theta
gradDesc(theta)
```

```python
def computeCost(x,y):
    a=(theta[0]+x*theta[1])-y
    b=x*a
    d = sum(np.power(a,2))/2
    return d
computeCost(x,y)
```

```python
e = []
f = []
for i in range (1,count+1,1):
    print(computeCost(x,y))
    e.append(computeCost(x,y))
    f.append(i)
    print(gradDesc(theta))
```
